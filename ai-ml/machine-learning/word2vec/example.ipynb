## What is Word2Vec?
Word2Vec is a technique that turns words into vectors (numbers) so that:

Similar words -> similar vectors
Relationships are preserved * Example: king - man + woman ≈ queen
This overall concept is known as word embeddings.

## Two Architectures (core idea)
CBOW (Continuous Bag of Words)
Skip-gram (more powerful)

# Example: CBOW (Continuous Bag of Words)
Predict target word from context

"The cat sits on the mat"
Context: ["The", "cat", "on", "the"]
Target: "sits"
# Example: Skip-gram (more powerful)
Target: "sits" Predict: ["The", "cat", "on", "the"]

```python
from gensim.models import Word2Vec
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

sentences = [
    ["king", "is", "a", "strong", "man"],
    ["queen", "is", "a", "wise", "woman"],
    ["boy", "is", "a", "young", "man"],
    ["girl", "is", "a", "young", "woman"],
    ["prince", "is", "a", "young", "king"],
    ["princess", "is", "a", "young", "queen"]
]

sentences = [
    [word for word in sentence if word not in stop_words] 
    for sentence in sentences
] * 1000

model = Word2Vec(sentences, vector_size=50, window=2, min_count=1, sg=1, epochs=500)
similarity = model.wv.similarity("king", "queen")
print(similarity)
```
